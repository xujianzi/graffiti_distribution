```python
import folium
import pandas as pd

# define the world map
world_map = folium.Map()

# display world map
world_map
```

```python
# NYC latitude and longitude values
latitude = 40.71
longitude = -73.94

# Create map and display it
nyc_map = folium.Map(location=[latitude, longitude], zoom_start=10)
nyc_map
```

```python
# Read Dataset 
url = 'http://xujianzi.github.io/data/graffiti.csv'
gdata = pd.read_csv(url)
gdata.head()
```

```python
## generate sample
sample1 = gdata.sample(n = 2000)
sample1.head()
```

```python
# get the first 2000 crimes in the cdata as the test data
limit = 2000
data = gdata.iloc[0:limit, :]

# Instantiate a feature group for the incidents in the dataframe
events = folium.map.FeatureGroup()

# Add events to the maps feature
for lat, lng in zip(data.latitude, data.longitude):
    events.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=7, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4
        )
    )

# Add events to map
nyc_map = folium.Map(location=[latitude, longitude], zoom_start=11)
nyc_map.add_child(events)
## save the file as html
#san_map.save(outfile= 'graffiti.html')
```
